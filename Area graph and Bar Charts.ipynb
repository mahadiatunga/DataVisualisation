```python
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df.head()
```

```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
df.head()
```

```python
# let's examine the types of the column labels
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str, df.columns))

# let's check the column labels types now
all(isinstance(column, str) for column in df.columns)
```

```python
df.set_index('Country', inplace=True)

# let's view the first five elements and see how the dataframe was changed
df.head()
```

```python
df['Total'] = df.sum(axis=1)

df.head()
```

```python
print ('data dimensions:', df.shape)
```

```python
years = list(map(str, range(1980, 2014)))

years
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
years2 = list(map(str, range(1985, 2011))) 

df1=df[df['Region'].str.contains('Africa')] #Extracting only African regions

df2 = df1.groupby(['Region']).sum() #summing to regional level from country wise values
```

```python
#df2.index look whatsthe new index
df2 = df2[years2].transpose() 
df2.index = df2.index.map(int) #Changing index to int for plotting
```

```python
#Graph Plotting
df2.plot(kind='area', 
             alpha=0.5, 
             stacked=False,
             figsize=(25, 10),
            )

plt.title('Immigration Trend of African Regions 1985 to 2010',fontsize=36)
plt.ylabel('Number of Immigrants',fontsize=26)
plt.xlabel('Years',fontsize=26)

plt.show()
```

```python
df3 = df.loc[['Greece','Albania','Bulgaria'], years2].transpose()
count,bin_edges = np.histogram(df3,15)

# un-stacked histogram
df3.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.35,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

```python
df4 = df.loc[['Denmark','Norway','Sweden'], years2].transpose()
count,bin_edges = np.histogram(df4,15)

# un-stacked histogram
df4.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.35,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

#### Qn. Draw area graph for different parts of Africa. Note you will use RegName

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

df_can.head()
```

```python
df=df_can.copy()
df2=df[(df.AreaName == 'Africa')]
df2
```

```python
df3=df2.groupby('RegName').sum()
df3
```

```python
df4=df3.drop(['AREA','DEV','REG'],axis='columns')
df4
```

```python
df4.T.plot.area()
plt.show()
```
